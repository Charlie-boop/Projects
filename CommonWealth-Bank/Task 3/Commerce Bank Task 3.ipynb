```python
import requests
import json

API_KEY = "7YHgax7SjaoCRDOmXjhpAnprC"
API_KEY_SECRET = "LP6ij2l0LttU642CqcA2GQvFSxy1LNklecdpWnB46iZVcm4cYS"
BEARER_TOKEN = "AAAAAAAAAAAAAAAAAAAAANlm2AEAAAAAx657crOsImazQXVxJqOefYAZj%2Bk%3DKlV48X0ymseOooRVipkVWUSa6Etod2sG5vJqOcb2cp5DtLrYF5"
handle = "CommBank"

# Function to get tweets with pagination
def get_tweets(handle, max_results=75, max_pages=20):
    all_tweets = []
    next_token = None
    page_count = 0
    
    while page_count < max_pages:
        # Base URL
        url = f"https://api.twitter.com/2/tweets/search/recent?query=from:{handle}&max_results={max_results}&tweet.fields=created_at&expansions=author_id&user.fields=created_at"
        
        # Add pagination token if we have one
        if next_token:
            url += f"&next_token={next_token}"
            
        headers = {"Authorization": f"Bearer {BEARER_TOKEN}"}
        
        response = requests.get(url, headers=headers)
        
        if response.status_code == 200:
            data = response.json()
            
            # Print the full JSON structure for the first page to examine it
            if page_count == 0:
                print("Full JSON response structure:")
                print(json.dumps(data, indent=4))
                print("\n" + "="*50 + "\n")
            
            # Get tweets from this page
            tweets = data.get('data', [])
            if tweets:
                all_tweets.extend(tweets)
                print(f"Retrieved {len(tweets)} tweets on page {page_count+1}")
            else:
                print(f"No tweets found on page {page_count+1}")
                break
            
            # Check if there's a next token for pagination
            meta = data.get('meta', {})
            next_token = meta.get('next_token')
            
            if not next_token:
                print("No more pages available")
                break
                
            page_count += 1
        else:
            print(f"Error: {response.status_code} - {response.text}")
            break
    
    return all_tweets

# Get tweets
tweets = get_tweets(handle)

# Print summary
print(f"\nTotal tweets retrieved: {len(tweets)}")

# Print the actual tweets
print("\nTweet content:")
for i, tweet in enumerate(tweets, 1):
   print(f"{i}. [{tweet.get('created_at')}] {tweet.get('text')[:100]}...")  # Fixed the parenthesis mismatch here
```
